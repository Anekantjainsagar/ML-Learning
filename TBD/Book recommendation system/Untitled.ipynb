```python
## Popularity Based
```

```python
books.head()
```

```python
ratings_new = ratings.merge(books, on='ISBN')
```

```python
ratings_new.head()
```

```python
ratings_new.shape
```

```python
num_of_ratings = ratings_new.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_of_ratings
```

```python
num_of_ratings.rename(columns={'Book-Rating': 'Num-Ratings'}, inplace=True)
```

```python
num_of_ratings
```

```python
avg_of_ratings = ratings_new.groupby('Book-Title')['Book-Rating'].mean().reset_index()
```

```python
avg_of_ratings.rename(columns={'Book-Rating': 'Avg-Ratings'}, inplace=True)
```

```python
avg_of_ratings
```

```python
df = avg_of_ratings.merge(num_of_ratings, on='Book-Title')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df = df[df['Num-Ratings'] >= 200].sort_values('Avg-Ratings', ascending=False).head(50)
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.merge(books, on='Book-Title').drop_duplicates('Book-Title')
```

```python
df.shape
```

```python
df.head()
```

```python
df = df[['Book-Title', 'Avg-Ratings', 'Num-Ratings', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
```

```python
## Collabrative filtering based data
```

```python
ratings_new.head()
```

```python
x = ratings_new.groupby('User-ID').count()['Book-Rating'] > 200
num_of_ratings_user = x[x].index
```

```python
num_of_ratings_user.shape
```

```python
filtered_ratings = ratings_new[ratings_new['User-ID'].isin(num_of_ratings_user)]
```

```python
y = ratings_new.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.shape
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns=['User-ID'], values='Book-Rating')
```

```python
pt.shape
```

```python
pt.fillna(0, inplace=True)
```
